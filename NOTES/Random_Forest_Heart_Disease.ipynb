**Import Libraries**

```python
import pandas as pd # Import the pandas library for data manipulation
import numpy as np # Import the numpy library for numerical operations

# Import train_test_split from sklearn.model_selection for splitting data into training and testing sets
from sklearn.model_selection import train_test_split

# Import RandomForestClassifier from sklearn.ensemble for building a Random Forest classification model
from sklearn.ensemble import RandomForestClassifier

# Import various metrics from sklearn.metrics for evaluating model performance
from sklearn.metrics import (
    accuracy_score, # For calculating the accuracy of the model
    confusion_matrix, # For creating a confusion matrix to evaluate classification performance
    classification_report # For generating a detailed classification report
)

# Import joblib for saving and loading Python objects, especially scikit-learn models
import joblib
```

**Load Dataset**

```python
df = pd.read_csv( # Read the CSV file 'heart.csv' into a pandas DataFrame
    "heart.csv"
)

df.head() # Display the first 5 rows of the DataFrame to get a glimpse of the data
```

**Check Dataset**

```python
df.shape # Display the number of rows and columns in the DataFrame
```

**Check missing values:**

```python
df.isnull().sum() # Check for the sum of null values in each column of the DataFrame
```

**Dataset information:**

```python
df.info() # Display a summary of the DataFrame, including data types and non-null values
```

**Separate Input and Output**

```python
# Independent Variables:

X = df.drop( # Create a new DataFrame 'X' by dropping the 'target' column from the original DataFrame
    "target", # Column to drop
    axis=1 # Specify that we are dropping a column (axis=1)
)
```

```python
X
```

```python
# Dependent Variable:

y = df["target"] # Create a Series 'y' containing only the 'target' column (dependent variable)
```

```python
y
```

y

**Split Dataset**

```python
X_train, X_test, y_train, y_test = train_test_split( # Split the data into training and testing sets

    X, # Independent variables
    y, # Dependent variable
    test_size=0.2, # Allocate 20% of the data for testing
    random_state=42 # Set a random seed for reproducibility

)
```

**Create Random Forest Model**

```python
model = RandomForestClassifier( # Initialize a RandomForestClassifier model

    n_estimators=100, # Number of decision trees in the forest
    random_state=42 # Set a random seed for reproducibility

)

# n_estimators=100  -> Create 100 Decision Trees
```

**Train Model**

```python
model.fit( # Train the Random Forest model
    X_train, # Training data for independent variables
    y_train # Training data for the dependent variable
)
```

**Prediction**

```python
y_pred = model.predict( # Make predictions on the test set
    X_test # Test data for independent variables
)
```

### **Evaluation**

**Accuracy**

```python
accuracy = accuracy_score( # Calculate the accuracy score
    y_test, # Actual target values for the test set
    y_pred # Predicted target values for the test set
)

print( # Print the calculated accuracy
accuracy
)
```

**Confusion Matrix**

```python
cm = confusion_matrix( # Generate the confusion matrix
    y_test, # Actual target values for the test set
    y_pred # Predicted target values for the test set
)

print(cm) # Print the confusion matrix
```

**Classification Report**

```python
print( # Print the classification report
classification_report( # Generate a classification report including precision, recall, f1-score
    y_test, # Actual target values for the test set
    y_pred # Predicted target values for the test set
)
)
```

**Feature Importance**

```python
importance = pd.DataFrame( # Create a DataFrame to store feature importances

{
"Feature":X.columns, # Column names as features
"Importance":model.feature_importances_ # Feature importance values from the trained model

}

)


importance.sort_values( # Sort the DataFrame by 'Importance' in descending order
    by="Importance", # Sort by this column
    ascending=False # Sort in descending order
)
```

**Save Model**

```python
joblib.dump( # Save the trained model to a file
    model, # The model object to save
    "heart_model.pkl" # The filename for the saved model
)
```
