**Import Libraries**

```python
import pandas as pd
import numpy as np

# Import modules for splitting data into training and testing sets
from sklearn.model_selection import train_test_split

# Import module for converting text into numerical features
from sklearn.feature_extraction.text import TfidfVectorizer

# Import the Logistic Regression model for classification
from sklearn.linear_model import LogisticRegression

# Import various metrics for model evaluation
from sklearn.metrics import (
accuracy_score,
confusion_matrix,
classification_report,
precision_score,
recall_score,
f1_score
)
```

**Load Dataset**

```python
import pandas as pd

# Load the dataset from a TSV file (tab-separated values)
df = pd.read_csv(
    "SMSSpamCollection",
    sep="\t",
    names=["label","message"]
)

# Display the first 5 rows of the DataFrame
print(df.head())

# Print the shape of the DataFrame (number of rows, number of columns)
print(df.shape)

# Count the occurrences of each unique value in the 'label' column
print(df["label"].value_counts())
```

**Check Dataset**

```python
# Display the first 5 rows of the DataFrame
df.head()
```

```python
# Print a concise summary of the DataFrame, including data types and non-null values
df.info()
```

```python
# Print the shape of the DataFrame (number of rows, number of columns)
df.shape
```

**Select Features**

```python
# Input feature: Assign the 'message' column to variable X
X = df["message"]
```

```python
# Output target: Assign the 'label' column to variable y
y = df["label"]
```

Convert Text into Numbers

```python
# Create a TF-IDF Vectorizer object
tfidf = TfidfVectorizer(
# Exclude common English stop words during tokenization
stop_words="english"
)
```

```python
import numpy as np

# Transform the text data (X) into TF-IDF features

# The error indicates 'X' might have been previously overwritten as a sparse matrix.
# Ensure 'X' refers to the original text messages from the DataFrame before transformation.
X = df["message"]

# Transform the text data (X) into TF-IDF features, and assign the result back to X
# so that subsequent steps like train_test_split use the vectorized data.
X = tfidf.fit_transform(X)

# If X_vectorized is needed as a separate variable, it can be assigned from the now-vectorized X.
X
```

**Split Dataset**

```python
# Split the dataset into training and testing sets
X_train,X_test,y_train,y_test=train_test_split(
# Input features
X,
# Target variable
y,
# Proportion of the dataset to include in the test split (20%)
test_size=0.2,
# Seed for random number generator for reproducibility
random_state=42
)
```

**Create Model**

```python
# Initialize a Logistic Regression model
model = LogisticRegression()
```

**Train Model**

```python
# Train the Logistic Regression model using the training data
model.fit(
X_train,
y_train
)
```

**Prediction**

```python
# Make predictions on the test set using the trained model
y_pred=model.predict(
X_test
)
```

**Evaluation Metrics**

**Confusion Matrix**

```python
# Calculate the confusion matrix for evaluation
confusion_matrix(
y_test,
y_pred
)
```

**Accuracy**

```python
# Calculate the accuracy score
# This measures the proportion of correctly classified instances

# Accuracy=(TP+TN)/TP+TN+FP+FN​
accuracy_score(
y_test,
y_pred
)
```

**Precision**

Out of predicted spam emails:

"How many were actually spam?"

Important when false alarms are costly.

```python
# Calculate the precision score
# Precision is the ratio of true positives to the sum of true positives and false positives.
# It answers: "Out of all predicted spam emails, how many were actually spam?"
# P=(TP)/(TP+FP)

precision_score(
y_test,
y_pred,
pos_label='spam'
)
```

**Recall**

```python
# Calculate the recall score
# Recall is the ratio of true positives to the sum of true positives and false negatives.
# It answers: "Out of all actual spam emails, how many were correctly identified?"
# R=(TP)/TP+FN

recall_score(
y_test,
y_pred,
pos_label='spam'
)
```

**F1 Score**

```python
# Calculate the F1 score
# The F1 score is the harmonic mean of precision and recall.
# It provides a single score that balances both precision and recall.
f1_score(
y_test,
y_pred,
pos_label='spam'
)
```

```python
# Generate a classification report, which includes precision, recall, f1-score, and support for each class
classification_report(
y_test,
y_pred
)
```

**Save Model**

```python
import joblib

# Save the trained Logistic Regression model to a file named 'spam_model.pkl'
joblib.dump(
model,
"spam_model.pkl"
)

# Save the fitted TF-IDF Vectorizer to a file named 'tfidf.pkl'
joblib.dump(
tfidf,
"tfidf.pkl"
)
```
