**Import Libraries**

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

**Download Dataset**

```python
df = yf.download("AAPL", start="2020-01-01", end="2025-01-01") # Download historical stock data for Apple (AAPL) from 2020-01-01 to 2025-01-01
```

```python
# Use only Closing Price
data = df[['Close']] # Extract only the 'Close' price column from the downloaded DataFrame
```

```python
data # Display the 'Close' price DataFrame
```

**Normalize Data**

```python
scaler = MinMaxScaler() # Initialize the MinMaxScaler to scale data between 0 and 1

scaled_data = scaler.fit_transform(data) # Apply min-max scaling to the 'Close' price data
```

**Create Sequences**

```python
X = [] # Initialize an empty list to store input sequences
y = [] # Initialize an empty list to store target values

# Loop through the scaled data to create sequences for the LSTM model
for i in range(60, len(scaled_data)): # Start from index 60 to ensure each sequence has 60 data points
    X.append(scaled_data[i-60:i,0]) # Append the past 60 data points as an input sequence
    y.append(scaled_data[i,0]) # Append the 61st data point as the target output

X = np.array(X) # Convert the list of input sequences to a NumPy array
y = np.array(y) # Convert the list of target values to a NumPy array

X = X.reshape((X.shape[0], X.shape[1], 1)) # Reshape the input array to be 3-dimensional for LSTM: (samples, timesteps, features)
```

**Build LSTM Model**

```python
model = Sequential() # Initialize a sequential model for building the neural network

# Add the first LSTM layer with 50 units, returning sequences for the next LSTM layer
model.add(LSTM(50, return_sequences=True, input_shape=(60,1)))

# Add the second LSTM layer with 50 units (does not return sequences as it's the last LSTM layer)
model.add(LSTM(50))

model.add(Dense(1)) # Add a Dense (fully connected) output layer with 1 unit for regression
```

**Loss Function**

```python
# Base Code

model.compile(

    optimizer="adam", # Configure the model with the Adam optimizer

    loss="mean_squared_error" # Set the loss function to Mean Squared Error (MSE)

)
```

**Backpropagation**

```python
history = model.fit(

    X, # Input features (sequences)

    y, # Target values

    epochs=20, # Train the model for 20 epochs (iterations over the entire dataset)

    batch_size=32 # Process data in batches of 32 samples at a time

)
```
